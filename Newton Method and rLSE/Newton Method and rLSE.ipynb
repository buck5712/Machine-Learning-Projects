```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python
def elimination(row1, row2, col, target=0):
    factor =((row2[col]-target)/row1[col])
    for i in range(len(row2)):
        row2[i] = row2[i]-row1[i]*factor

def gauss(A):
    for i in range(len(A)):
        if A[i][i] == 0:
            for j in range(i+1, len(A)):
                if A[i][j]!=0:
                    A[i], A[j] = A[j],A[i]
                    break
            else:
                raise ValueError("Matrix is not invertible!")
        for j in range(i+1, len(A)):
            elimination(A[i], A[j], i)
    for i in range(len(A)-1,-1,-1):
        for j in range(i-1, -1, -1):
            elimination(A[i], A[j], i)
    for i in range(len(A)):
        elimination(A[i], A[i], i, target=1)
    return A
            

def inverse(A):
    #生成一個單位矩陣
    n = len(A)
    I = np.zeros((n,n))
    for i in range(n):
        I[i,i]=1 
    #concate A及I
    A_I = copy.deepcopy(A)
    A_I = np.concatenate((A_I, I), axis=1)
    #對concate後的矩陣做高斯消去
    gauss(A_I)
    #取A_I右半部為A之反矩陣
    A_inverse = A_I[:, (A_I.shape[1]//2):]
    return A_inverse

def Transpose(matrix):
    matrix = np.array(matrix)
    matrix_m = matrix.shape[0]
    matrix_n = matrix.shape[1]
    result = []
    for i in range(matrix_n):
        result.append(tuple([matrix[j][i] for j in range(matrix_m)]))
    return result


def MatrixMultiplication(matrix1, matrix2):
    result = []
    for row in matrix1:
        result_row = []
        for col in Transpose(matrix2):
            result_row.append(sum([i * j for (i, j) in zip(row, col)]))
        result.append(tuple(result_row))
    return result

def cal_error(error_matrix):
    error = sum([error[0] * error[0] for error in error_matrix])
    return error
```

```python
def rLSE(obs_x, obs_y, n, lamda):
    X_design_matrix = np.expand_dims(np.power(obs_x, n-1), axis=1)
    for i in range(n-2, -1, -1):
        X_design_matrix = np.append(X_design_matrix, np.expand_dims(np.power(obs_x, i), axis=1), axis=1)
    obs_y = np.expand_dims(obs_y, axis=1)
    X_design_matrix_T = Transpose(X_design_matrix)
    
    #簡化sol計算寫法
    ATA = MatrixMultiplication(X_design_matrix_T, X_design_matrix)
    ATA_plus_Lamda_I = ATA + np.identity(len(X_design_matrix_T))*lamda
    ATb = MatrixMultiplication(X_design_matrix_T, obs_y)
    rLSE_sol = MatrixMultiplication(inverse(ATA_plus_Lamda_I),ATb)
    
    fitting_line = ""
    for i in range(len(rLSE_sol)-1, -1, -1):
        if i==0:
            fitting_line += str(rLSE_sol[len(rLSE_sol)-1-i][0])
        else:
            fitting_line += str(rLSE_sol[len(rLSE_sol)-1-i][0]) + "X^" + str(i) + " + "
    
    error_matrix = MatrixMultiplication(X_design_matrix, rLSE_sol)-obs_y
    error = cal_error(error_matrix)
    
    return rLSE_sol, fitting_line, error
```

```python
def Newton(obs_x, obs_y, n, lamda):
    X_design_matrix = np.expand_dims(np.power(obs_x, n-1), axis=1)
    for i in range(n-2, -1, -1):
        X_design_matrix = np.append(X_design_matrix, np.expand_dims(np.power(obs_x, i), axis=1), axis=1)
    obs_y = np.expand_dims(obs_y, axis=1)
    X_design_matrix_T = Transpose(X_design_matrix)
    
    #簡化sol計算寫法
    ATA = MatrixMultiplication(X_design_matrix_T, X_design_matrix)
    ATb = MatrixMultiplication(X_design_matrix_T, obs_y)
    Newton_sol =  MatrixMultiplication(inverse(ATA), ATb)    
    fitting_line = ""
    for i in range(len(Newton_sol)-1, -1, -1):
        if i==0:
            fitting_line += str(Newton_sol[len(Newton_sol)-1-i][0])
        else:
            fitting_line += str(Newton_sol[len(Newton_sol)-1-i][0]) + "X^" + str(i) + " + "
    

    error_matrix = MatrixMultiplication(X_design_matrix, Newton_sol)-obs_y
    error = cal_error(error_matrix)
    
    return Newton_sol, fitting_line, error
```

```python
def draw(x_data, y_data, sol):
    sol = np.squeeze(sol)
    polynomial = np.poly1d(sol)
    x_axis = np.linspace(-6, 6)
    y_axis = polynomial(x_axis)
    fig = plt.figure()
    axes = fig.add_axes([0.1, 0.1, 2, 1])
    axes.plot(x_axis, y_axis)
    axes.plot(x_data, y_data, '.')
    plt.show()
```

```python
df = pd.read_csv("testfile.txt", header=None)
df.columns = ['obs_x', 'obs_y']
obs_x, obs_y = df['obs_x'], df['obs_y']
```

```python
n = [2, 3, 3]
lamda = [0, 0, 10000]
for j in range(3):
    print("Case {}: n = {},lamda = {}".format(j+1, n[j], lamda[j]))
    sol_rLSE, fitting_line, cost = rLSE(obs_x, obs_y, n[j], lamda[j])
    print("LSE:")
    print("Fitting line: {}".format(fitting_line))
    print("Total error: {}".format(cost))
    #print("\n")
    sol_Newton, fitting_line, cost = Newton(obs_x, obs_y, n[j], lamda[j])
    print("Newton's Method:")
    print("Fitting line: {}".format(fitting_line))
    print("Total error: {}".format(cost))
    draw(obs_x, obs_y, sol_rLSE)
    draw(obs_x, obs_y, sol_Newton)
    print("\n")
```
