```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from numpy.linalg import inv, det, cholesky
import csv
```

```python
def load_data(path = 'data/input.data'):
    x=[]
    y=[]
    file = open(path,'r')
    for line in file.readlines():
        datapoint = line.split(' ')
        x.append(float(datapoint[0]))
        y.append(float(datapoint[1]))
    
    file.close()
    x = np.asarray(x)
    y = np.asarray(y)
    return x,y

def kernel(x1, x2, sigma=1, alpha=1, length_scale=1):
    kernel = sigma**2 * np.power(1+(np.power(x1.reshape(-1,1)-x2.reshape(1,-1),2.0))/(2*alpha*length_scale**2), -alpha)
    
    return kernel

def predict(x_star, x, y, K, beta, alpha=1, length_scale=1):
    
    K_x_xstar = kernel(x, x_star, alpha=1, length_scale=1)
    K_xstar_xstar = kernel(x_star, x_star, alpha=1, length_scale=1)
    
    means = K_x_xstar.T @ np.linalg.pinv(K) @y.reshape(-1,1)
    var = K_xstar_xstar + np.identity(len(K_xstar_xstar))*(1/beta) - K_x_xstar.T @ np.linalg.pinv(K) @ K_x_xstar
    
    return means, var
     
def NegativeLogLikelihood(theta, x, y, beta):
    theta = theta.reshape(-1)
    K = kernel(x, x, theta[0], theta[1], theta[2])
    K += np.identity(len(x), dtype=np.float64) * (1 / beta)
    nll = 0.5 * np.log(det(K))
    nll += 0.5 * y.T @ inv(K) @ y
    nll += 0.5 * len(x) * np.log(2 * np.pi)
    return nll
```

```python
x, y = load_data()
beta = 5
K = kernel(x, x, alpha=1, length_scale=1, sigma=1)
K = K + np.identity(len(x))*(1/beta)
print(K)
```

```python
x_star=np.linspace(-60,60,num=500)
mean_predict,variance_predict=predict(x_star,x,y,K,beta,alpha=1,length_scale=1)
mean_predict=mean_predict.reshape(-1)
sigma_predict=np.sqrt(np.abs(np.diag(variance_predict)))
print(sigma_predict)
```

```python
plt.plot(x,y,'bo')
plt.plot(x_star,mean_predict,'k-')
plt.fill_between(
    x_star,mean_predict+1.96*sigma_predict,mean_predict-1.96*sigma_predict,facecolor='salmon')
plt.xlim(-60,60)
plt.show()
```

```python
sigma = 1
alpha = 1
length_scale = 1
```

```python
method = ['SLSQP', 'BFGS', 'L-BFGS-B']
for m in method:

    opt = minimize(NegativeLogLikelihood, [sigma, alpha, length_scale], 
                        bounds=((1e-8, 1e6), (1e-8, 1e6), (1e-8, 1e6)), 
                        args=(x, y, beta),method = m )
    sigma_opt = opt.x[0]
    alpha_opt = opt.x[1]
    length_scale_opt = opt.x[2]
    print("Minimize Slover:" , m)
    print("sigma_opt :", sigma_opt)
    print("alpha_opt : ", alpha_opt)
    print("length_scale_opt :", length_scale_opt)
    print(" ")
    
    # do Gaussian process again
    K = kernel(x, x, alpha=alpha_opt, length_scale=length_scale_opt)
    K = K + np.identity(len(x), dtype=np.float64) * (1 / beta)
    K_inv = np.linalg.pinv(K)
    x_star = np.linspace(-60,60,num=500)
    mean_predict, variance_predict = \
        predict(x_star, x, y, K, beta, alpha=alpha_opt, length_scale=length_scale_opt)
    mean_predict = mean_predict.reshape(-1)
    #print(np.diag(variance_predict))
    #variance_predict[variance_predict < 0] = 0
    sigma_predict = np.sqrt(np.abs(np.diag(variance_predict)))
    #print(sigma_predict)
    #plot
    plt.plot(x,y,'bo')
    plt.plot(x_star,mean_predict,'k-')
    plt.fill_between(x_star,mean_predict+1.96*sigma_predict,\
        mean_predict-1.96*sigma_predict,facecolor='salmon')
    plt.xlim(-60,60)
    plt.title('minimize slover: {}'.format(m))
    plt.show()
```
